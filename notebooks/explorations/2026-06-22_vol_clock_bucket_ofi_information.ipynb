# Volatility Clock Bucket OFI Information Analysis

This notebook reorganizes the bucketed OFI data into per-asset volatility bars. A volatility bar closes when cumulative squared 5-minute return reaches roughly six average 5-minute bars of realized variance for that asset.

This is diagnostic event time, not a causal live clock: the boundary uses realized variance inside the bar. The goal is to test whether flow information looks stronger after normalizing for quiet versus volatile periods.

Signals are low-bucket OFI, high-bucket OFI, and high-minus-low OFI spread. Targets are same-vol-bar and forward vol-clock returns.

## Working Conclusions

- Volatility clocks reduce the dominance of calendar-time regime changes, but they do not turn size-bucket OFI into a strong standalone forward predictor in this sample.
- Same-vol-bar relationships remain stronger for high-size bucket OFI than low-size bucket OFI, consistent with large trades carrying immediate price-impact information.
- Forward vol-clock horizons still show weak and inconsistent predictive relationships; nominal t-stats appear in isolated assets/features and should be treated as hypothesis candidates, not evidence of edge.
- The OFI spread can be useful diagnostically because it asks whether large trades are more one-sided than small trades in the same volatility unit, but it is not uniformly superior to high-bucket OFI.
