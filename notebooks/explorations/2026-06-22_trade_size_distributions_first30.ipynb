# Trade Size Distributions by Aggressor Side

Exploratory inspection of trade-size distributions for the first 30 active GLBX roots used in the macro futures work. This notebook is an artifact browser: it does not rerun Databento ingestion.

**Source:** cached raw trade parquet files from `research/databento-asset-browser`.

**Window:** `2026-05-17` to `2026-06-16`; outright contracts only; RTH `13:30-20:00 UTC`.

**Side convention:** `B` is buy aggressor and `A` is sell aggressor.

Each plot has a side-normalized trade-size histogram on top and the tail curve `P(size >= x)` on bottom.

```python
from __future__ import annotations

from pathlib import Path

import pandas as pd
from IPython.display import display

ASSET_DIR = Path("notebooks/explorations/assets/2026-06-22_trade_size_distributions_first30")
if not ASSET_DIR.exists():
    ASSET_DIR = Path("assets/2026-06-22_trade_size_distributions_first30")

summary = pd.read_csv(ASSET_DIR / "trade_size_distribution_summary.csv")
display(summary.head(12))
```

```python
pivot = summary.pivot(index=["root", "asset"], columns="side", values="p99").reset_index()
pivot["buy_minus_sell_p99"] = pivot["B"] - pivot["A"]
display(pivot.sort_values("buy_minus_sell_p99", key=lambda s: s.abs(), ascending=False).head(15))
```
