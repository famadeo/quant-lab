# Bucket OFI Information Analysis

This notebook tests whether bottom-size or top-size bucket order-flow imbalance carries more information about returns.

Universe: first 30 active GLBX roots. Window: `2026-05-17` to `2026-06-16`, RTH `13:30-20:00 UTC`.

Return targets are the existing continuous 5-minute return series. `same_bar` measures contemporaneous price impact. Forward horizons use returns after the OFI bar: next 5, 15, 30, and 60 minutes.

Regressions are simple `return_bps = alpha + beta * bucket_OFI` with Newey-West t-stats using 12 lags. This is exploratory and not multiple-testing adjusted.

## Working Conclusions

- Same-bar relationships are much stronger than forward relationships. That is expected: signed aggressive flow and returns are mechanically linked within the bar.
- Forward relationships decay quickly and are not stable enough in this 21-trading-day sample to claim a robust edge.
- Top-size bucket OFI is not uniformly more informative than bottom-size bucket OFI. The answer depends on asset class and horizon.
- In rates, top-size bucket OFI often captures most volume and has cleaner same-bar impact, but forward predictiveness remains weak.
- In equity/energy contracts, the bottom bucket is usually one-lot dominated and can show microstructure/bid-ask-bounce behavior; treat it as a liquidity/noise diagnostic before treating it as informed flow.
- The next step for a tradable claim would be a walk-forward test using pre-registered assets/horizons, costs, and controls for lagged return/reversal.
