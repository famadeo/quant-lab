# Trade Size Bucket OFI

This notebook compares order-flow imbalance for the bottom and top trade-size buckets for the first 30 active GLBX roots.

**Window:** `2026-05-17` to `2026-06-16`; outright contracts only; RTH `13:30-20:00 UTC`; sides `B`/`A` only.

**Bucket rule:** bottom bucket is `size <= q10`; top bucket is `size >= q90`; ties at thresholds are included.

**OFI formula:** `sum(signed_size) / sum(size)` where `B=+size` and `A=-size`. The high-minus-low spread is `OFI_high - OFI_low`.

Raw 5-minute OFI observations are plotted as faint points; bold lines are `20`-bar rolling means. CVD panels are cumulative signed size within each bucket.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display

ASSET_DIR = Path("notebooks/explorations/assets/2026-06-22_trade_size_bucket_ofi")
if not ASSET_DIR.exists():
    ASSET_DIR = Path("assets/2026-06-22_trade_size_bucket_ofi")

summary = pd.read_csv(ASSET_DIR / "trade_size_bucket_ofi_summary.csv")
display(summary)
```
