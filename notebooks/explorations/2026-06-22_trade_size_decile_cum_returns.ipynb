# Trade Size Decile Cumulative Log Returns

This notebook inspects whether the smallest and largest trade-size buckets line up with different cumulative price-move profiles. It is exploratory diagnostics, not a trading strategy or PnL backtest.

**Universe:** first 30 active GLBX roots.

**Window:** `2026-05-17` to `2026-06-16`; outright contracts only; RTH `13:30-20:00 UTC`; sides `B`/`A` only.

**Method:** for each root, compute q10/q90 of trade size across RTH trades. The bottom bucket is `size <= q10`; the top bucket is `size >= q90`; ties at thresholds are included. For each selected trade, compute `log(price_t) - log(previous price in the same contract)`, using the full-session same-contract predecessor before RTH filtering. Then sum those increments into 5-minute bins and cumulative-sum by bucket.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display

ASSET_DIR = Path("notebooks/explorations/assets/2026-06-22_trade_size_decile_cum_returns")
if not ASSET_DIR.exists():
    ASSET_DIR = Path("assets/2026-06-22_trade_size_decile_cum_returns")

summary = pd.read_csv(ASSET_DIR / "trade_size_decile_cum_return_summary.csv")
display(summary)
```

```python
cols = [
    "root",
    "asset",
    "q10_size",
    "q90_size",
    "bottom_share",
    "top_share",
    "final_bottom_cum_logret_pct",
    "final_top_cum_logret_pct",
    "final_top_minus_bottom_pct",
]
display(
    summary[cols]
    .sort_values("final_top_minus_bottom_pct", key=lambda s: s.abs(), ascending=False)
    .head(15)
)
```
